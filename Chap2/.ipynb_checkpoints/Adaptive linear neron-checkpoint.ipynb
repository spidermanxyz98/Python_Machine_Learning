```python
import numpy as np
```

```python
class AdalineGD(object):
    
    def __init__(self, eta=0.0001, n_iter=50):
        self.eta=eta
        self.n_iter=n_iter
        
    def fit(self, X, y):
        self.w=np.zeros(X.shape[1]+1)
        self.cost=[]
        
        for _ in range(self.n_iter):
            error=y-self.net_input(X)
            self.w[1:]+=self.eta*X.T.dot(error)
            self.w[0]+=self.eta*sum(error)
            
            self.cost.append(error.dot(error)/2)
            
        return self
            
    def net_input(self, X):
        return np.dot(X,self.w[1:])+self.w[0]
    
    def activation(self,X):
        return self.net_input(X)
    
    def prediction(self, X):
        return np.where(self.activation(X)>=0.0, 1, -1)
```

```python
import pandas as pd
%matplotlib inline

df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
y=df.iloc[0:100, 4].values
y=np.where(y=='Iris-setosa',-1,1)
X=df.iloc[0:100, [0,2]].values
X[:,0]=(X[:,0]-X[:,0].mean())/X[:,0].std()
X[:,1]=(X[:,1]-X[:,1].mean())/X[:,1].std()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[y==1,0], X[y==1,1], color='blue', marker='o', label='category1')
plt.scatter(X[y==-1][:,0], X[y==-1][:,1], color="red", marker='x', label='category2')
plt.xlabel('petal length')
plt.ylabel('sepal length')
plt.legend(loc='upper left')
plt.show()
```

```python
model=AdalineGD(eta=0.01, n_iter=15)
model.fit(X,y)
y_pre=model.prediction(X)
```

```python
plt.scatter(range(model.n_iter),model.cost)
plt.xlabel('number of iteration')
plt.ylabel('cost function')
plt.title('cost function-number of iteration')
```

```python
resolution=0.02
x1_min, x1_max=X[:,0].min()-1, X[:,0].max()+1
x2_min, x2_max=X[:,1].min()-1, X[:,1].max()+1
xx1, xx2=np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
Z=model.prediction(np.array([xx1.ravel(),xx2.ravel()]).T)
Z=Z.reshape(xx1.shape)

plt.contourf(xx1, xx2, Z, alpha=0.3)
plt.scatter(X[y==1,0], X[y==1,1], color='red', alpha=0.8)
plt.scatter(X[y==-1,0], X[y==-1,1], color='blue', alpha=0.8)
plt.xlabel('feature1')
plt.ylabel('feature2')
```
